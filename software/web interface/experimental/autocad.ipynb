```python
import cairosvg
from PIL import Image
import matplotlib.pyplot as plt
import io

# Convert SVG to PNG (or another raster format)
svg_file = "floor_plan_1.svg"
png_data = cairosvg.svg2png(url=svg_file)

# Load PNG into an image object
image = Image.open(io.BytesIO(png_data))

# Display with Matplotlib
plt.figure(figsize=(15, 15))
plt.imshow(image)
plt.axis('off')
# plt.title("Rendered SVG")
plt.show()
```
